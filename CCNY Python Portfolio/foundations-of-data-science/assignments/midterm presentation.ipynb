```python
ChileEDF['Obligations (Constant Dollars)'].quantile([0.25,0.75])
```

```python
ChileEDF['Obligations (Constant Dollars)'].describe()
```

```python
mean_C = ChileEDF['Obligations (Constant Dollars)'].mean()
std_C = ChileEDF['Obligations (Constant Dollars)'].std()

lower_bound = mean_C - 3 * std_C
upper_bound = mean_C + 3 * std_C

FChileEDF = ChileEDF[(ChileEDF['Obligations (Constant Dollars)'] >= lower_bound) & (ChileEDF['Obligations (Constant Dollars)'] <= upper_bound)]
```

```python
FChileEDF.set_index('Fiscal Year', inplace=True)
FChileEDF.reset_index(inplace=True)
plt.scatter(FChileEDF['Fiscal Year'], FChileEDF['Obligations (Constant Dollars)'], color='blue')
plt.title('US Economic Aid to Chile (no outliers)')
plt.xlabel('Fiscal Year')
plt.ylabel('Obligations (Constant Dollars)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
